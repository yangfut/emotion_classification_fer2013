```
data = pd.read_csv('/content/drive/My Drive/fer2013/fer2013.csv')
print(data)
pix = data['pixels']
for i in range (pix.shape[0]):
  data['pixels'][i] = np.asarray(pix[i].split(' ')).astype(float).reshape(48,48) / 255
  data['pixels'][i] = np.round(data['pixels'][i],8)
print(data)
data.to_pickle('/content/drive/My Drive/fer2013/fer2013.pkl')
```

```
#@title pixs
def PlotImg(x_train):
  plt.figure(1,figsize = (20,20))
  img = x_train.reshape(x_train.shape[0],x_train.shape[1],x_train.shape[2])
  pics = 100
  for i in range(0,pics):
    plt.subplot(5,20,i+1)
    plt.imshow(img[i],cmap='Greys',interpolation='nearest')
    plt.axis('off')
pixs = np.stack(data['pixels'], axis = -3)
print(type(pixs))
print(pixs[0:10])
PlotImg(pixs)
```
